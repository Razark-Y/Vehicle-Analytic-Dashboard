```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
car_data = pd.read_csv('car_prices.csv')
```

```python
car_data.info()
```

```python
car_data.dropna(subset=['saledate'], inplace=True)
car_data.dropna(subset=['vin'],inplace=True)
car_data['make'] = car_data['make'].fillna('Other')
car_data['interior'] = car_data['interior'].replace("—", None)
car_data['color'] = car_data['color'].replace("—", None)
car_data['model'] = car_data['model'].fillna('Other')
car_data['trim'] = car_data['trim'].fillna('Other')
car_data['color'] = car_data['color'].fillna('Other')
car_data['body'] = car_data['body'].fillna(car_data['body'].mode()[0])
car_data['transmission'] = car_data['transmission'].fillna(car_data['transmission'].mode()[0])
car_data['interior'] = car_data['interior'].fillna(car_data['interior'].mode()[0])
car_data['condition'].fillna(car_data['condition'].median(), inplace=True)
car_data['odometer'].fillna(car_data['odometer'].mean(), inplace=True)
car_data['mmr'].fillna(car_data['mmr'].mean(), inplace=True)
```

```python
car_data.isnull().sum()
```

```python
def parse_date(date_str):
    try:
        return datetime.strptime(date_str.strip(), "%a %b %d %Y %H:%M:%S GMT%z")
    except ValueError:
        return None  
car_data['saledate'] = car_data['saledate'].str.replace(r"\(.*\)", "", regex=True)
car_data['saledate'] =car_data['saledate'].apply(parse_date)
```

```python
car_data['state'] = car_data['state'].str.upper()
```
